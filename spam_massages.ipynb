```python
import tensorflow as tf
import numpy as np
```

```python
messages = [
    "win free cash now click link",
    "congratulations you won prize",
    "urgent verify your account password",
    "aap free mobile jeet gaye hain link par click karein",
    "hello are we meeting today",
    "please send me assignment",
    "kal class 10 baje hogi",
    "mujhe project file bhej dena"
]

labels = np.array([
    1, 1, 1, 1,   # spam
    0, 0, 0, 0    # no spam
])
```

```python
vectorizer = tf.keras.layers.TextVectorization(
    output_mode="int",
    output_sequence_length=20
)

vectorizer.adapt(messages)
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,), dtype=tf.string),
    vectorizer,
    tf.keras.layers.Embedding(1000, 16),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(8, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(
    tf.convert_to_tensor(messages, dtype=tf.string),
    labels,
    epochs=30,
    verbose=1
)
```

```python
model.fit(
    tf.convert_to_tensor(messages, dtype=tf.string),
    labels,
    epochs=30,
    verbose=1
)

loss, accuracy = model.evaluate(tf.convert_to_tensor(messages, dtype=tf.string), labels, verbose=0)

print("Accuracy:", round(float(accuracy) * 100, 2), "%")
```
